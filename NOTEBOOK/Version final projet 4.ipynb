## 2eme question Nombre théorique de personne qui pourrait être nourries

l'apport en calories moyen pour un être humain est de 2400 calories par jours

```python
dispo_alimentaire_population_2017 = pd.merge(population_filtered_2017,dispo_alimentaire,on='Zone',how='left')
```

```python
dispo_alimentaire_population_2017.head()
```

```python
dispo_alimentaire_population_2017['Disponibilité alimentaire (Kcal)'] = dispo_alimentaire_population_2017['Disponibilité alimentaire (Kcal/personne/jour)'] * dispo_alimentaire_population_2017['population']*365
dispo_alimentaire_population_2017.head()
```

```python
dispo_alimentaire_par_pays_2017 = dispo_alimentaire_population_2017.groupby('Zone').sum()
dispo_alimentaire_par_pays_2017.head()
```

```python
dispo_alimentaire_par_pays_2017.reset_index(inplace=True)
dispo_alimentaire_global_2017 = dispo_alimentaire_par_pays_2017[['Zone','Disponibilité alimentaire (Kcal)']]
dispo_alimentaire_global_2017.head()
```

```python
total_Kcal = dispo_alimentaire_global_2017['Disponibilité alimentaire (Kcal)'].sum()
total_Kcal
```

```python
nombre_theorique_personnes_nourries = round(total_Kcal / 2400)
nombre_theorique_personnes_nourries
```
